```python
#¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.
sns.scatterplot(data=df, x="infant_mortality", y="life_expectancy")
plt.xlabel('Mortalidad Infantil')
plt.ylabel('Esperanza de Vida')
plt.title('Relación esperanza de vida y mortalidad infantil')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
```

```python
# Configuración de la figura: 2 filas y 2 columnas
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Relación entre Tasa de Natalidad y Variables Socio-Demográficas', fontsize=18)

# 1. Natalidad vs Mortalidad Infantil
sns.scatterplot(ax=axes[0, 0], data=df, x='infant_mortality', y='birth_rate', color='coral')
axes[0, 0].set_title('Natalidad vs Mortalidad Infantil')

# 2. Natalidad vs Esperanza de Vida
sns.scatterplot(ax=axes[0, 1], data=df, x='life_expectancy', y='birth_rate', color='skyblue')
axes[0, 1].set_title('Natalidad vs Esperanza de Vida')

# 3. Natalidad vs Densidad de Población
sns.scatterplot(ax=axes[1, 0], data=df, x='density', y='birth_rate', color='green')
axes[1, 0].set_title('Natalidad vs Densidad de Población')

# 4. Natalidad vs Población Urbana (%)
sns.scatterplot(ax=axes[1, 1], data=df, x='urban_population', y='birth_rate', color='purple')
axes[1, 1].set_title('Natalidad vs Población Urbana')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
